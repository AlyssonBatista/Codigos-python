# Biblioteca matplotlib


 



```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
data = np.linspace(-5,5)
plt.plot(data)
plt.show()
```

```python
t = np.arange(0,2,0.01)
signal = 1 + np.sin(2 * np.pi * t)

plt.plot(t,signal)
plt.show()
```

```python
t = np.arange(0,2,0.05)
signal = 1 + np.sin( 2 * np.pi * t)

plt.plot(t-1,signal,'r--',label = 'sinal modificado')
plt.plot(t,signal,label =" Sinal original")
plt.legend(loc = 1)

plt.title('Funções de exemplo')
plt.xlabel('Tempo(s)')
plt.ylabel('Tensão(mV)')

plt.show()
```

```python
t = np.arange(0,2,0.01)
signal1 = 1 + np.sin(2 * np.pi * t)
signal2 = 1 + np.cos(2 * np.pi * t)

plt.plot(t,signal1,label="Seno",c="b")
plt.plot(t,signal2,label="Cosseno",c="orange")
plt.legend(loc= 1)

plt.title("Funções de exemplo")
plt.ylabel("Amplitude")
plt.xlabel("t(s)")

plt.plot(t,signal1,t,signal2)
plt.show()
```

```python
t = np.arange(0,2,0.05)
signal = 1 + np.sin(2 * np.pi * t)

plt.plot(t,signal,label="Sinal original")

plt.plot(t-1,signal,'ro',label = 'Sinal modificado')
plt.legend(loc = 1)

plt.title('Funções exemplo')
plt.xlabel('tempo(s)')
plt.ylabel('tensão(mV)')

plt.show()
```

```python
plt.suptitle('Título geral')

plt.subplot(221)
plt.plot(t,signal)

plt.subplot(222)
plt.plot(t,signal*2)

plt.subplot(223)
plt.plot(t,signal/2)

plt.subplot(224)
plt.plot(t,signal + signal * 10)
```

```python
x = np.linspace(0.1, 2 * np.pi , 41)
y = np.exp(np.sin(x))

plt.stem(x,y,use_line_collection = True)
plt.show()
```
