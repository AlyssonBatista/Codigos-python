```python
import numpy as np
import matplotlib.pyplot as plt


n=np.arange(-3,4)
x=np.arange(-3,4)

plt.style.use('Solarize_Light2')
plt.stem(n,x,use_line_collection=True)
plt.title("x[n]")
plt.show()
```

## Impulson initário δ(t)

```python
def impseq(n0,n1,n2):
    n = np.arange(n1,n2+1)
    x = np.where(n-n0==0,1,0)
    return [n,x]

n = impseq(0,-5,5)
x = impseq(0,-5,5)

plt.stem(n[0],x[1],use_line_collection=True)
plt.title("δ[n]")
plt.show()
```

## Degrau unitário

```python
def stepseq(n0,n1,n2):
    if n0<n1 or n0>n2 or n1>n2:
        raise ValueError("é necessário satisfazer n1 <= n0 <= n2")
    else:
        n = np.arange(n1,n2+1)
        x = np.where(n-n0 >= 0,1,0)
    return [n,x]

n = stepseq(0,-10,10)
x = stepseq(0,-10,10)
plt.stem(n[0],x[1],use_line_collection=True)
plt.title('u[n]')
plt.show()
```

## Sinais senoidais

### Gerar os gráficos para a sequência discreta x[n] =2·cos(πn/6), -20 ≤ n ≤ 20

```python
from math import pi

n = np.arange(-20,20+1)

x = 2 * np.cos((pi*n)/6)

plt.xticks(n[::4])
plt.yticks(n[::1])
plt.stem(n,x,use_line_collection=True)
plt.show()
```

### Trace o gráfico das seguintes sequências discretas e determine se estes sinais são periódicos.

1. x[n] = cos(3πn/8), 0 ≤ n ≤ 30
2. x[n] = cos(n), 0 ≤ n ≤ 30
3. x[n] = 5·cos(0.5πn + π/4) + 2·sin(0.5πn), 0 ≤ n ≤ 30

 #### x[n] = cos(3πn/8), 0 ≤ n ≤ 30

```python
n = np.arange(0,31)

x = np.cos((3*np.pi*n)/8)


plt.stem(n,x,use_line_collection=True)
plt.title("x[n] = cos(3πn/8)")
plt.show()
```

#### x[n] = cos(n), 0 ≤ n ≤ 30

```python
n = np.arange(0,31)

x = np.cos(n)

plt.stem(n,x,use_line_collection=True)
plt.title("x[n] = cos(n)")
plt.show()
```

#### x[n] = 5·cos(0.5πn + π/4) + 2·sin(0.5πn), 0 ≤ n ≤ 30

```python
n = np.arange(0,31)

x = 5*np.cos(0.5*np.pi*n + (np.pi)/4) + 2 * np.sin(0.5 * np.pi * n)

plt.stem(n,x,use_line_collection=True)
plt.title("x[n] = 5·cos(0.5πn + π/4) + 2·sin(0.5πn)")
plt.show()
```

## Convolução

```python
x = np.array([0,1,2,3,2,1])
g = np.array([1,1,1,1,1,1])

conv = np.convolve(x,g)
n = np.arange(0,x.size+g.size-1)

plt.stem(n,conv,use_line_collection=True)
plt.title("x1[n] = x[n]*g[n]")
plt.show()
```
